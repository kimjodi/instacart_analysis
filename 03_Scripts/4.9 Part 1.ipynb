## 04. Importing df_ords_prods_merged data and conducting data wrangling (prepping to be merged)

```python
# Importing ords_prods_merge data
df_ords_prods_merge = pd.read_pickle(os.path.join(path, '02_Data', 'Prepared_Data', 'orders_products_merged_latest.pkl'))
```

```python
# Checking ords_prods_merge
df_ords_prods_merge.head(20) 
```

```python
# Dropping 'order_spending_flag' column as it's a duplicate of 'order_frequency_flag' column
df_ords_prods_merge = df_ords_prods_merge.drop(columns = ['order_spending_flag'])
```

```python
# Dropping 'busiest_day' column as it's been replaced with 'busiest_days' column
df_ords_prods_merge = df_ords_prods_merge.drop(columns = ['busiest_days'])
```

```python
# Dropping 'busiest_day' column as it's been replaced with 'busiest_days' column
df_ords_prods_merge = df_ords_prods_merge.drop(columns = ['busiest_day'])
```

```python
# Adding back 'busiest_days' as it was accidently removed':(
labeling_counts = df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
labeling_counts.index
```

```python
# As index in this case is day nummber and sorted (ascending), 2 busiest days and 2 slowest days are found

day_number = list(labeling_counts.index)
```

```python
day_number
```

```python
results = []
```

```python
for value in df_ords_prods_merge['orders_day_of_week']:
    if value == day_number[0] or value == day_number[1]:
        results.append('Busiest days')
    elif value == day_number[len(day_number) - 1] or value == day_number[len(day_number) - 2]:
        results.append('Slowest days')
    else:
        results.append('Regulary busy days')
```

```python
df_ords_prods_merge['orders_day_of_week'].shape
```

```python
df_ords_prods_merge['busiest_days'] = results
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```
