#### Q2.In this Exercise, you learned how to find the aggregated mean of the “order_number” column grouped by “department_id” for a subset of your dataframe. Now, repeat this process for the entire dataframe.

```python
# Aggregating data 'oder_number' and 'department_id'

df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

#### Q4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Creating a new column 'max_order' from 'user_id' and 'order_number' columns

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking new column

df_ords_prods_merge.head(15)
```

### 04. Creating loyalty flag

```python
# Creating a loyalty flag criteria: (loyal customer >=40), (regular customer <=40 - >10), (new customer <=<10)

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
df_ords_prods_merge['loyalty_flag'].value_counts(dropna= False)
```

#### Q6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria: If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Creating a new column 'user_spending' from 'user_id' and 'prices' columns

df_ords_prods_merge['user_spending'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Checking new column

df_ords_prods_merge.head(15)
```

### 05. Creating a spending flag

```python
# Creating a spending flag: lower than 10, then flagged as a “Low spender", higher than or equal to 10, then flagged as a “High spender"

df_ords_prods_merge.loc[df_ords_prods_merge['user_spending'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['user_spending'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
df_ords_prods_merge['spending_flag'].value_counts(dropna= False)
```

```python
df_ords_prods_merge.head(10)
```

#### Q7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows: If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.” If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.” If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
df_ords_prods_merge.groupby('days_since_prior_order')
```

```python
df_ords_prods_merge.groupby('user_id').agg({'days_since_prior_order': ['mean', 'min', 'max']})
```

```python
df_ords_prods_merge.groupby('user_id').agg({'days_since_prior_order': ['median']})
```

```python
# Creating a new column 'median_prior_order' from 'user_id' and 'days_since_prior_order' columns

df_ords_prods_merge['median_prior_order'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Checking new column

df_ords_prods_merge.head(10)
```

```python
# Creating new order spending flag

df_ords_prods_merge.loc[df_ords_prods_merge['median_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent Customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent Customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_prior_order'] > 10) & (df_ords_prods_merge['median_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular Customer'
```

```python
# Checking output

df_ords_prods_merge['order_frequency_flag'].value_counts(dropna= False)
```
